```python
import numpy as np
import skimage as ski
from skimage import io, color, exposure, img_as_float
import matplotlib.pyplot as plt
from scipy import signal
```

#Edge detection

```python
# Load square.png
I = io.imread('/content/square.png')

# Normalize the image
I = img_as_float(I)


I = I[:,:,3]

plt.imshow(I, cmap = 'gray')
plt.axis("off")
plt.show()

print(I.shape)
```

Implement these filters -

sobel_G.png

```python
# Define the filter here
g_x = np.array([[-1, 0, 1],
                [-2, 0, 2],
                [-1, 0, 1]])

g_y = np.array([[1,  2,  1],
                [0,  0,  0],
                [-1, -2, -1]])


print("Gx - \n", g_x)
print("\nGy -\n", g_y)
```

```python
# Apply these kernel on the image
# Use cross correlation operation
# Ensure the image size remain same after filtering
I_x = signal.convolve2d(I, g_x, boundary='symm', mode='same')
I_y = signal.convolve2d(I, g_y, boundary='symm', mode='same')

fig = plt.figure(figsize=(14, 8))

ax1 = fig.add_subplot(1,2,1)
ax1.imshow(I_x, cmap = 'gray')
ax1.title.set_text("Filtered with Gx")
ax1.axis("off")

ax2 = fig.add_subplot(1,2,2)
ax2.imshow(I_y, cmap = 'gray')
ax2.title.set_text("Filtered with Gy")
ax2.axis("off")

plt.show()

print("Ix shape - ", I_x.shape)
print("Iy shape - ", I_y.shape)
```

```python
# Apply the kernels again on the original image but using convolution operation
# Ensure image dimensions are preserved
I_x_conv = signal.convolve2d(I, np.flipud(np.fliplr(g_x)), mode='same', boundary='symm')
I_y_conv = signal.convolve2d(I, np.flipud(np.fliplr(g_y)), mode='same', boundary='symm')


fig = plt.figure(figsize=(14, 8))

ax1 = fig.add_subplot(2,2,1)
ax1.imshow(I_x, cmap = 'gray')
ax1.title.set_text("Ix correlated")
ax1.axis("off")

ax2 = fig.add_subplot(2,2,2)
ax2.imshow(I_x_conv, cmap = 'gray')
ax2.title.set_text("Ix convolved")
ax2.axis("off")


ax3 = fig.add_subplot(2,2,3)
ax3.imshow(I_y, cmap = 'gray')
ax3.title.set_text("Iy correlated")
ax3.axis("off")


ax4 = fig.add_subplot(2,2,4)
ax4.imshow(I_y_conv, cmap = 'gray')
ax4.title.set_text("Iy convolved")
ax4.axis("off")

plt.show()

print("Ix (convolved) shape - ", I_x_conv.shape)
print("Iy (convolved) shape - ", I_y_conv.shape)
```

# Localized histeq

```python
# Load the "land.jpeg" image
I = io.imread('/content/land.jpeg')

# Normalize the image using img_to_float
I = img_as_float(I)


# Calculate the histogram
hist, bins = np.histogram(exposure.equalize_adapthist(I, clip_limit=0.03).flatten(), bins=256, range=[0, 1])


# Calculate the cumulative histogram
chist, cbins = np.cumsum(hist) / hist.sum() * 100 , bins[:-1]




plt.rcParams['figure.figsize'] = (16, 8)
plt.subplot(2, 1, 1)
plt.imshow(I, cmap = 'gray', vmin=0, vmax=1)
plt.axis("off")
plt.title("image")

plt.subplot(2, 2, 3)
plt.plot(bins, hist, 'k')
plt.xlabel("pixel values")
plt.title("histogram")

plt.subplot(2, 2, 4)
plt.plot(cbins,chist, 'r', lw=3)
plt.ylabel("percentage")
plt.title("cumulative histogram")

plt.show()
```

```python
# Perform global histogram equalization on the image
I_histeq = exposure.equalize_hist(I)

# Calculate the histogram of the globally equalized image
hist_I_histeq, bins_I_histeq = np.histogram(I_histeq.flatten(), bins=256, range=[0, 1])

# Perform adaptive histogram equalization (AHE) with kernel size = (41, 41)
I_ahe = exposure.equalize_adapthist(I, kernel_size=(41, 41), clip_limit=1.0)  # clip_limit=1 means no contrast limiting

# Calculate the histogram of the AHE equalized image
hist_I_ahe, bins_I_ahe = np.histogram(I_ahe.flatten(), bins=256, range=[0, 1])

# Perform CLAHE with kernel size = (41, 41) and clip limit = 0.05
I_clahe = exposure.equalize_adapthist(I, kernel_size=(41, 41), clip_limit=0.05)

# Calculate the histogram of the CLAHE equalized image
hist_I_clahe, bins_I_clahe = np.histogram(I_clahe.flatten(), bins=256, range=[0, 1])
```

```python
plt.rcParams['figure.figsize'] = (18,6)
plt.subplot(2, 4, 1)
plt.imshow(I, cmap = 'gray', vmin=0, vmax=1)
plt.axis("off")
plt.title("I")

plt.subplot(2, 4, 2)
plt.imshow(I_histeq, cmap = 'gray', vmin=0, vmax=1)
plt.axis("off")
plt.title("I_histeq")

plt.subplot(2, 4, 3)
plt.imshow(I_ahe, cmap = 'gray', vmin=0, vmax=1)
plt.axis("off")
plt.title("I_ahe")

plt.subplot(2, 4, 4)
plt.imshow(I_clahe, cmap = 'gray', vmin=0, vmax=1)
plt.axis("off")
plt.title("I_clahe")

plt.subplot(2, 4, 5)
plt.plot(bins[:-1], hist, 'k')
plt.xlabel("pixel values")
plt.title("hist(I)")

plt.subplot(2, 4,6)
plt.plot(bins_I_histeq[:-1],hist_I_histeq , 'k')
plt.xlabel("pixel values")
plt.title("hist(I_histeq) ")

plt.subplot(2, 4, 7)
plt.plot(bins_I_ahe[:-1] ,hist_I_ahe, 'k')
plt.xlabel("pixel values")
plt.title("hist(I_ahe) ")

plt.subplot(2, 4, 8)
plt.plot(bins_I_clahe[:-1],hist_I_clahe, 'k')
plt.xlabel("pixel values")
plt.title("hist(I_clahe)")

plt.show()
```
